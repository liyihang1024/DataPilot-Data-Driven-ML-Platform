```python
import pandas as pd
from scipy import stats
import numpy as np

def detect_outliers(df):
    """
    检测异常值并返回一个包含异常值信息的DataFrame。
    """
    outliers = []

    for col in df.select_dtypes(include=np.number).columns:
        z_score = np.abs(stats.zscore(df[col].dropna()))  # 排除NaN值
        outliers_indices = df.index[(z_score > 3)]  # 通常Z-score大于3认为是异常值
        outliers.extend([(index, col) for index in outliers_indices])

    # 如果检测到异常值，返回包含它们的DataFrame
    if outliers:
        outliers_df = pd.DataFrame(outliers, columns=['Index', 'Column'])
        return outliers_df
    else:
        return pd.DataFrame(columns=['Index', 'Column'])

def fill_outliers(df, method='mean'):
    """
    处理异常值。
    可选方法包括 'mean', 'median', 'mode', 'drop_row', 'drop_col'
    """
    for col in df.select_dtypes(include=np.number).columns:
        z_score = np.abs(stats.zscore(df[col]))
        outlier_indexes = df[z_score > 3].index
        
        if method == 'mean':
            df.loc[outlier_indexes, col] = df[col].mean()
        elif method == 'median':
            df.loc[outlier_indexes, col] = df[col].median()
        elif method == 'mode':
            df.loc[outlier_indexes, col] = df[col].mode()[0]
        elif method == 'drop_row':
            df.drop(index=outlier_indexes, inplace=True)
        elif method == 'drop_col':
            df.drop(columns=col, inplace=True)
    return df

# 读取数据
data = pd.read_excel('./test-data.xlsx', header=0, index_col=0)

# 检测异常值
outliers = detect_outliers(data)
print("检测到的异常值:\n", outliers)

# 处理异常值
# 这里可以选择 'mean', 'median', 'mode', 'drop_row', 'drop_col' 其中的一个
method = 'mean'
data_clean = fill_outliers(data, method)
print(f"\n异常值处理后的数据（使用{method}方法）:\n", data_clean.head())
```
